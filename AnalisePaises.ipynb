```
#Importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```
#renomeando colunas
df.rename(columns={
'country':'País',
'Region':'Região',
'Surface area (km2)':'Área em Km2',
'Population in thousands (2017)':'População em milhares',
'Population density (per km2, 2017)':'Densidade pop. por km2',
'Sex ratio (m per 100 f, 2017)':'% de homens p/ 100 mulheres',
'GDP: Gross domestic product (million current US$)':'PIB em USD',
'GDP per capita (current US$)':'PIB per capita em USD',
'Economy: Agriculture (% of GVA)':'Valor agregado Agricultura',
'Economy: Industry (% of GVA)':'Valor agregado Industria',
'Economy: Services and other activity (% of GVA)':'Valor agregado Serviços',
'Unemployment (% of labour force)':'Desemprego',
'Urban population (% of total population)_x': 'População urbana (% do Total)',
'Urban population growth rate (average annual %)':'Crescimento população urbana (Média anual)',
'Population age distribution (0-14 / 60+ years, %)':'Idade (0-14 / 60+, %)',
'Infant mortality rate (per 1000 live births':'Mortalidade infantil',
'Health: Total expenditure (% of GDP)':'Gasto com saúde (% do PIB)',
'Health: Physicians (per 1000 pop.)':'Médicos por 1000 hab.',
'Education: Government expenditure (% of GDP)':'Gasto com educação (% do PIB)',
'CO2 emission estimates (million tons/tons per capita)':'Emissões CO2',
'Pop. using improved drinking water (urban/rural, %)':'Acesso a água potável (urbano/rural, %)',
'Pop. using improved sanitation facilities (urban/rural, %)':'Acesso saneamento básico (urbano/rural, %)',
'Quality Of Life Index':'Qualidade de vida',
'Purchasing Power Index':'Poder de compra',
'Safety Index':'Indice de segurança',
'Health Care Index':'Indice de saude',
'Cost Of Living Index':'Custo de vida',
'Consumer price index (2010 = 100)':'Inflação desde 2010',
'Current health expenditure (% of GDP)':'Gasto com saude (% do PIB)',
'Life expectancy at birth, female (years)':'Expectativa de vida mulheres',
'Life expectancy at birth, male (years)':'Expectativa de vida homens',
'Life expectancy at birth, total (years)':'Expectativa de vida geral',
'Population, female':'População mulheres',
'Population, male':'População homens',
'Tax revenue (% of GDP)':'Arrecadação de impostos (% do PIB)',
'Taxes on income, profits and capital gains (% of revenue)': 'Impostos sobre renda, lucros e ganhos de capital (% da arrecadalção)'
}, inplace=True)
```

```
#Verificando os tipos de dados
df.dtypes
```

```
#Traduzindo regições
list(df['Região'].unique())
df.replace({'SouthAmerica':'America do Sul', 
            'WesternEurope':'Europa Ocidental', 
            'EasternEurope':'Europa Oriental',
            'SouthernEurope':'Europa Meridional',
            'NorthernAmerica':'America do Norte',
            'EasternAsia':'Ásia Oriental',
            'WesternAsia':'Ásia Ocidental',
            'NorthernEurope':'Europa Setentrional',
            'NorthernAfrica':'Africa Setentrional',
            'SouthernAsia':'Ásia Meridional',
            'South-easternAsia':'Sudeste Asiático',
            'CentralAmerica':'America Central',
            'SouthernAfrica':'Africa Meridional',
            }, inplace=True)

list(df['Região'].unique())
```

```
#Visualizando as 5 primeiras linhas
df.head()
```

```
#Describe do Data Frame
df.describe()
```

```
#Ajusta formato da população para milhões 
df['População em milhares'] = df['População em milhares']/1000
df.rename(columns={'População em milhares':'População em milhões'}, inplace=True)
```

```

#Adiciona colunas % Idosos, % Crianças
df[['% Crianças', '% Idosos']] = df['Idade (0-14 / 60+, %)'].str.split("/", expand=True)
#remove coluna original
df.drop(["Idade (0-14 / 60+, %)"], axis=1, inplace=True)
```

```
#Ajusta pib para bilhões

df['PIB em USD'] = df['PIB em USD']/1000
df.head()
```

 **Missing Values**

```
#Verificando se temos dados faltantes
df.isnull().sum()
```

**PIB per capita - Regiões**

```
pd.options.display.float_format = '{:20,.2f}'.format
#Média de pib per capita por região
df_media = df.groupby(df["Região"])["PIB per capita em USD"].agg(["mean"]).sort_values(['mean'], ascending=False).reset_index()
df_media.rename(columns={"mean":"Média PIB per capita"}).style.format({'Média PIB per capita':'${0:,.2f}'}).set_properties(**{'text-align': 'center'}).hide_index()
```

**TOP 30 Países por renda per capita**

```
#Resetando o index
df_top30 = df[['País', 'PIB per capita em USD']].sort_values(by=['PIB per capita em USD'],ascending=False).head(30)
df_top30.style.format({'PIB per capita em USD':'${0:,.2f}'}).set_properties(**{'text-align': 'center'}).hide_index()
```
